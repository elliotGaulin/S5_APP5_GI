```python
n = 5
x = 2

x_i = np.arange(0, n+1)
p_3x8 = 1 / 64
p_4x5 = 1/ 125

combinations = [math.comb(n, x_i[k]) for k in range(len(x_i))]
f_x_3x8 = [combinations[k] * (p_3x8 ** x_i[k]) * ((1 - p_3x8) ** (n - x_i[k])) for k in range(len(x_i))]
f_x_4x5 = [combinations[k] * (p_4x5 ** x_i[k]) * ((1 - p_4x5) ** (n - x_i[k])) for k in range(len(x_i))]


p_x_beq_2_3x8 = np.sum(f_x_3x8[x:n])
p_x_beq_2_4x5 = np.sum(f_x_4x5[x:n])

mu_3x8 = n * p_3x8
mu_4x5 = n * p_4x5

var_3x8 = n * p_3x8 * (1 - p_3x8)
var_4x5 = n * p_4x5 * (1 - p_4x5)
```

```python
display(md("### Résultats"))
display(md("#### 3x8"))
display(md(f"- $P(X >= 2)= {p_x_beq_2_3x8:.4f}$"))
display(md(f"- $\\mu= {mu_3x8:.4f}$"))
display(md(f"- $\\sigma^2= {var_3x8:.4f}$"))
display(md("#### 4x5"))
display(md(f"- $P(X >= 2)= {p_x_beq_2_4x5:.4f}$"))
display(md(f"- $\\mu= {mu_4x5:.4f}$"))
display(md(f"- $\\sigma^2= {var_4x5:.4f}$"))
```
